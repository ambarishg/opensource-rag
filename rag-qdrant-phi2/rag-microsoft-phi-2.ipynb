```python
# Qdrant server URL
URL ="20.119.32.168"
# Qdrant dimension of the collection
DIMENSION = 384
# Qdrant collection name
COLLECTION_NAME = "SEARCH_ENGINE"
METRIC_NAME ="COSINE"
MODEL_NAME="all-MiniLM-L6-v2"
```

```python
import qdrant_client as qc
import qdrant_client.http.models as qmodels
from qdrant_client.http.models import *
```

```python
from sentence_transformers import SentenceTransformer
```

```python
import torch
from transformers import pipeline
```

```python
user_input = "Describe Platyhelminthes"
query_filter =  None
```

```python
def get_qdrant_client():
    client = qc.QdrantClient(url=URL)
    METRIC = qmodels.Distance.COSINE
    return client
```

```python
client = get_qdrant_client()
```

```python
client
```

```python
def load_model():
    return SentenceTransformer(MODEL_NAME)
```

```python
model = load_model()
xq = model.encode(user_input,convert_to_tensor=True)
```

```python
xq
```

```python
search_result = client.search(collection_name=COLLECTION_NAME,
                                    query_vector=xq.tolist(),
                                    query_filter=query_filter,
                                    limit=5)
```

```python
search_result
```

```python
contexts =""
for result in search_result:
    contexts +=  result.payload['token']+ "\n\n"
```

```python
contexts
```

```python
%%time
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

torch.set_default_device("cuda")

model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2", torch_dtype="auto", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2", trust_remote_code=True)
```

```python
prompt = f"""Instruct:Answer based on context:\n\n{contexts}\n\n{user_input}.\nOutput:"""
model_inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False)
```

```python
outputs = model.generate(**model_inputs, max_length=1000)
text = tokenizer.batch_decode(outputs)[0]
print(text)
```
